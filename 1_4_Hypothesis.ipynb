## –®–∞–≥ 5. –ü–æ–¥—Å—á—ë—Ç –≤—ã—Ä—É—á–∫–∏

–°–æ–∑–¥–∞–¥–∏–º –æ–±—â—É—é —Ç–∞–±–ª–∏—Ü—É —Å —Å—É–º–º–∞—Ä–Ω—ã–º —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ–º, —Å—É–º–º–∞—Ä–Ω—ã–º –≤—Ä–µ–º–µ–Ω–µ–º –ø–æ–µ–∑–¥–æ–∫ –∏ —á–∏—Å–ª–æ–º —Ç–∞–∫–∏—Ö –ø–æ–µ–∑–¥–æ–∫ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è. –î–æ–±–∞–≤–∏–º —Å—Ç–æ–ª–±–µ—Ü —Å –æ–∫—Ä—É–≥–ª–µ–Ω–Ω—ã–º–∏ –≤ –±–æ–ª—å—à—É—é —Å—Ç–æ—Ä–æ–Ω—É –¥–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—è–º–∏. –¢–∞–∫–∂–µ —Å–æ—Ö—Ä–∞–Ω–∏–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ —Ç–∞—Ä–∏—Ñ–Ω–æ–º –ø–ª–∞–Ω–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è

```python
aggregated_data = (
    data
    .assign(rounded_duration=lambda x: np.ceil(x['duration']))
    .groupby(['user_id', 'month'])
    .agg(
        month_distance=('distance', 'sum'),
        month_duration=('duration', 'sum'),
        month_rounded_duration=('rounded_duration', 'sum'),
        number_of_month_trips=('duration', 'count'),
        minute_price=('minute_price', 'first'),
        start_ride_price=('start_ride_price', 'first'),
        subscription_fee=('subscription_fee', 'first'),
        subscription_type=('subscription_type', 'first')
        )
    .reset_index()
)
```

```python
aggregated_data.head()
```

–í—ã—á–∏—Å–ª–∏–º –≤—ã—Ä—É—á–∫—É. –ò –¥–ª—è –Ω–∞–≥–ª—è–¥–Ω–æ—Å—Ç–∏ —É–¥–∞–ª–∏–º –Ω–µ–Ω—É–∂–Ω—ã–µ —Å—Ç–æ–ª–±—Ü—ã

```python
aggregated_data['revenue'] = (
    aggregated_data['start_ride_price']
    * aggregated_data['number_of_month_trips']
    + aggregated_data['minute_price']
    * aggregated_data['month_rounded_duration']
    + aggregated_data['subscription_fee']
)
```

 –í–∞–∂–Ω–æ –ø–æ–Ω—è—Ç—å, —Ç—Ä–∞—Ç—è—Ç –ª–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ —Å –ø–æ–¥–ø–∏—Å–∫–æ–π –±–æ–ª—å—à–µ –≤—Ä–µ–º–µ–Ω–∏ –Ω–∞ –ø–æ–µ–∑–¥–∫–∏. –ï—Å–ª–∏ –¥–∞, —Ç–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ —Å –ø–æ–¥–ø–∏—Å–∫–æ–π –º–æ–≥—É—Ç –±—ã—Ç—å ¬´–≤—ã–≥–æ–¥–Ω–µ–µ¬ª –¥–ª—è –∫–æ–º–ø–∞–Ω–∏–∏.

–°—Ñ–æ—Ä–º—É–ª–∏—Ä—É–µ–º –≥–∏–ø–æ—Ç–µ–∑—ã

$H_0$: –û–±–µ –≥—Ä—É–ø–ø—ã —Ç—Ä–∞—Ç—è—Ç –æ–¥–∏–Ω–∞–∫–æ–≤–æ–µ –≤—Ä–µ–º—è

$H_1$: –õ—é–¥–∏ —Å –ø–æ–¥–ø–∏—Å–∫–æ–π —Ç—Ä–∞—Ç—è—Ç –±–æ–ª—å—à–µ –≤—Ä–µ–º–µ–Ω–∏

<div class="alert alert-success" style="border-radius: 10px; box-shadow: 2px 2px 2px; border: 1px solid; padding: 10px ">
<b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ v.1</b> 
    
üëç 
–ì–∏–ø–æ—Ç–µ–∑—ã —Å—Ñ–æ—Ä–º—É–ª–∏—Ä–æ–≤–∞–Ω—ã –≤–µ—Ä–Ω–æ.
</div>

–ü—Ä–æ–≤–µ—Ä–∏–º –Ω–æ—Ä–º–∞–ª—å–Ω–æ –ª–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –¥–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –ø–æ–µ–∑–¥–æ–∫ –≤ –æ–±–µ–∏—Ö –≥—Ä—É–ø–ø–∞—Ö. –ü—Ä–æ–≤–µ–¥–µ–º [—Ç–µ—Å—Ç –®–∞–ø–∏—Ä–æ ‚Äî –£–∏–ª–∫–∞](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.shapiro.html).

```python
shapiro_ultra_duration = st.shapiro(data_ultra.duration)
shapiro_free_duration = st.shapiro(data_free.duration)

print(f'''–¢–µ—Å—Ç –®–∞–ø–∏—Ä–æ-–£–∏–ª–∫–∞ –¥–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å –ø–æ–¥–ø–∏—Å–∫–æ–π:\
 p-value = {shapiro_ultra_duration.pvalue}''')
print(f'–¢–µ—Å—Ç –®–∞–ø–∏—Ä–æ-–£–∏–ª–∫–∞ –¥–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –±–µ–∑ –ø–æ–¥–ø–∏—Å–∫–∏:\
 p-value = {shapiro_free_duration.pvalue}');
```

–ú—ã –ø–æ–ª—É—á–∏–ª–∏ –æ—á–µ–Ω—å –Ω–∏–∑–∫–æ–µ –∑–Ω–∞—á–Ω–∏–µ `p-value`, —Ç–æ –µ—Å—Ç—å –≥–∏–ø–æ—Ç–µ–∑—É –æ –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –º–æ–∂–µ–º –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å. –¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å T-test –Ω–µ –æ—á–µ–Ω—å –≤–µ—Ä–Ω–æ. –¢–µ–º –Ω–µ –º–µ–Ω–µ–µ –ø–æ–ø—Ä–æ–±—É–µ–º –ø—Ä–æ–≤–µ—Å—Ç–∏ –µ–≥–æ + –ø—Ä–æ–≤–µ–¥–µ–º –Ω–µ–ø–∞—Ä–∞–º–µ—Ç—Ä–∏—á–µ—Å–∫–∏–π —Ç–µ—Å—Ç –ú–∞–Ω–Ω–∞-–£–∏—Ç–Ω–∏, –∫–æ—Ç–æ—Ä—ã–π –±–æ–ª–µ–µ –ø—Ä–µ–¥–ø–æ—á—Ç–∏—Ç–µ–ª—å–Ω—ã–π –≤ –Ω–∞—à–µ–π –∑–∞–¥–∞—á–µ. 

**T-test**

```python
t_test_duration = st.ttest_ind(
    data_ultra.duration, 
    data_free.duration,
    alternative='greater'
    )

alpha = 0.05

print(f'–ù–µ–∑–∞–≤–∏—Å–∏–º—ã–π t-test: p-value = {t_test_duration.pvalue}')

if t_test_duration.pvalue < alpha:
    print('–ì–∏–ø–æ—Ç–µ–∑–∞ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è')
else:
    print('–ì–∏–ø–æ—Ç–µ–∑–∞ –Ω–µ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è')
```

> –ó–∞–º–µ—Ç–∏–º, —á—Ç–æ –µ—Å–ª–∏ –±—ã –¥–∞–Ω–Ω—ã–µ –≤—Å–µ-—Ç–∞–∫–∏ –±—ã–ª–∏ –Ω–æ—Ä–º–∞–ª—å–Ω–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã, —Ç–æ —Å—Ç–æ–∏–ª–æ –±—ã –ø—Ä–æ–≤–µ—Ä–∏—Ç—å —Ä–∞–≤–µ–Ω—Å—Ç–≤–æ –¥–∏—Å–ø–µ—Ä—Å–∏—è, —Ç–æ –µ—Å—Ç—å –ø—Ä–æ–≤–µ—Å—Ç–∏ –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ –ø—Ä–æ–≤–µ—Å—Ç–∏ [—Ç–µ—Å—Ç –õ–µ–≤–µ–Ω–∞](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.levene.html). –ò –ø–æ –µ–≥–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º —É–∫–∞–∑–∞–ª–∏ –±—ã –∑–Ω–∞—á–µ–Ω–∏–µ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π `equal_var`

**–¢–µ—Å—Ç –ú–∞–Ω–Ω–∞-–£–∏—Ç–Ω–∏**

```python
u_test_duration = st.mannwhitneyu(
    data_ultra.duration,
    data_free.duration,
    alternative='greater'
)

alpha = 0.05

print(f'U-test: p-value = {u_test_duration.pvalue}')

if u_test_duration.pvalue < alpha:
    print('–ì–∏–ø–æ—Ç–µ–∑–∞ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è')
else:
    print('–ì–∏–ø–æ—Ç–µ–∑–∞ –Ω–µ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è')
```

**–°—Ä–∞–≤–Ω–µ–Ω–∏–µ —Å—Ä–µ–¥–Ω–∏—Ö**

```python
print(f'''–°—Ä–µ–¥–Ω—è—è –∏ –º–µ–¥–∏–∞–Ω–Ω–∞—è –¥–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å –¥–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å –ø–æ–¥–ø–∏—Å–∫–æ–π:
    mean: {data_ultra.duration.mean()}
    median: {data_ultra.duration.median()}''')

print(f'''\n–°—Ä–µ–¥–Ω—è—è –∏ –º–µ–¥–∏–∞–Ω–Ω–∞—è –¥–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å –¥–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –±–µ–∑ –ø–æ–¥–ø–∏—Å–∫–∏:
    mean: {data_free.duration.mean()}
    median: {data_free.duration.median()}''')
```

**–í—ã–≤–æ–¥:**

–û–±–∞ —Ç–µ—Å—Ç–∞ –æ—Ç–≤–µ—Ä–≥–∞—é—Ç –≥–∏–ø–æ—Ç–µ–∑—É –æ —Ä–∞–≤–µ–Ω—Å—Ç–≤–µ —Å—Ä–µ–¥–Ω–∏—Ö –¥–≤—É—Ö –≤—ã–±–æ—Ä–æ–∫, –ø—Ä–∏ —ç—Ç–æ–º –ø—Ä–∏–Ω–∏–º–∞–µ—Ç—Å—è –≥–∏–ø–æ—Ç–µ–∑–∞ –æ —Ç–æ–º, —á—Ç–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ —Å –ø–æ–¥–ø–∏—Å–∫–æ–π —Ç—Ä–∞—Ç—è—Ç –±–æ–ª—å—à–µ –≤—Ä–µ–º–µ–Ω–∏ –Ω–∞ –ø–æ–µ–∑–¥–∫–∏. –í —Ü–µ–ª–æ–º, –º—ã –º–æ–∂–µ–º –Ω–∞–±–ª—é–¥–∞—Ç—å —ç—Ç–æ, —Å—Ä–∞–≤–Ω–∏–≤–∞—è —Å—Ä–¥–µ–Ω–∏–µ –∏ –º–µ–¥–∏–∞–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –¥–≤—É—Ö –≤—ã–±–æ—Ä–æ–∫

### –ì–∏–ø–æ—Ç–µ–∑–∞ –æ –¥–∏—Å—Ç–∞–Ω—Ü–∏–∏ –ø–æ–µ–∑–¥–∫–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å –ø–æ–¥–ø–∏—Å–∫–æ–π

–†–∞—Å—Å—Ç–æ—è–Ω–∏–µ –æ–¥–Ω–æ–π –ø–æ–µ–∑–¥–∫–∏ –≤ 3130 –º–µ—Ç—Ä–æ–≤ ‚Äî –æ–ø—Ç–∏–º–∞–ª—å–Ω–æ–µ —Å —Ç–æ—á–∫–∏ –∑—Ä–µ–Ω–∏—è –∏–∑–Ω–æ—Å–∞ —Å–∞–º–æ–∫–∞—Ç–∞. –°—Ç–æ–∏—Ç –ø—Ä–æ–≤–µ—Ä–∏—Ç—å –≥–∏–ø–æ—Ç–µ–∑—É –æ —Ç–æ–º, —á—Ç–æ —Å—Ä–µ–¥–Ω–µ–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ, –∫–æ—Ç–æ—Ä–æ–µ –ø—Ä–æ–µ–∑–∂–∞—é—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ —Å –ø–æ–¥–ø–∏—Å–∫–æ–π –∑–∞ –æ–¥–Ω—É –ø–æ–µ–∑–¥–∫—É, –Ω–µ –ø—Ä–µ–≤—ã—à–∞–µ—Ç 3130 –º–µ—Ç—Ä–æ–≤.

–°—Ñ–æ—Ä–º—É–ª–∏—Ä—É–µ–º –≥–∏–ø–æ—Ç–µ–∑—ã

 $H_0$: –°—Ä–µ–¥–Ω–µ–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å –ø–æ–¥–ø–∏—Å–∫–æ–π –Ω–µ –±–æ–ª—å—à–µ 3130 –º

 $H_1$: –ü–æ–¥–ø–∏—Å—á–∏–∫–∏ –≤ —Å—Ä–µ–¥–Ω–µ–º –ø—Ä–æ–µ–∑–∂–∞—é—Ç —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –±–æ–ª—å—à–µ 3130 –º

–¢–∞–∫–∂–µ –ø—Ä–æ–≤–µ—Ä–∏–º –Ω–æ—Ä–º–∞–ª—å–Ω–æ –ª–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –¥–∏—Å—Ç–∞–Ω—Ü–∏–∏ –ø–æ–µ–∑–¥–æ–∫

```python
shapiro_ultra_distance = st.shapiro(data_ultra.distance)

print(f'''–¢–µ—Å—Ç –®–∞–ø–∏—Ä–æ-–£–∏–ª–∫–∞ –¥–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å –ø–æ–¥–ø–∏—Å–∫–æ–π:\
 p-value = {shapiro_ultra_distance.pvalue}''');
```

–ú—ã –ø–æ–ª—É—á–∏–ª–∏ –æ—á–µ–Ω—å –Ω–∏–∑–∫–æ–µ –∑–Ω–∞—á–Ω–∏–µ `p-value`, —Ç–æ –µ—Å—Ç—å –≥–∏–ø–æ—Ç–µ–∑—É –æ –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –º–æ–∂–µ–º –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å. –í —Ü–µ–ª–æ–º, –º–æ–∂–Ω–æ –±—ã–ª–æ –ø–æ–Ω—è—Ç—å —ç—Ç–æ –∏ –ø–æ –≥—Ä–∞—Ñ–∏–∫—É —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è, –Ω–æ —Ç–µ–ø–µ—Ä—å –º—ã –≤ —ç—Ç–æ–º —Ç–æ—á–Ω–æ —É–≤–µ—Ä–µ–Ω—ã.

–ü—Ä–æ–≤–µ—Ä–∏–º —Å–∏–º–º–µ—Ç—Ä–∏—á–Ω–æ—Å—Ç—å –¥–∞–Ω–Ω—ã—Ö

```python
data_ultra.distance.skew()
```

–ó–¥–µ—Å—å —Å—Ç–æ–∏—Ç, –∫–∞–∫ –∏ –¥–ª—è –ø—Ä–æ—à–ª–æ–π –≥–∏–ø–æ—Ç–µ–∑—ã, –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –Ω–µ–ø–∞—Ä–∞–º–µ—Ç—Ä–∏—á–µ—Å–∫–∏–π —Ç–µ—Å—Ç. –û–¥–Ω–∞–∫–æ, —Å—É–¥—è –ø–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—é, –¥–∏—Å—Ç–∞–Ω—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –Ω–µ —Ç–æ–ª—å–∫–æ –Ω–µ –Ω–æ—Ä–º–∞–ª—å–Ω–æ, –Ω–æ –∏ –∞—Å—Å–∏–º–µ—Ç—Ä–∏—á–Ω–æ (—Å—É–¥—è –ø–æ –ø–æ–∫–∞–∑–∞—Ç–µ–ª—é –ø–µ—Ä–µ–∫–æ—Å–∞ $\in[-1, -0.5]$, –º—ã –∏–º–µ–µ–º —É–º–µ—Ä–µ–Ω–Ω–æ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–π (–ª–µ–≤—ã–π) –ø–µ—Ä–µ–∫–æ—Å). –¢–∞–∫ —á—Ç–æ –∏ –Ω–µ–ø–∞—Ä–∞–º—Ç–µ—Ä–∏—á–µ—Å–∫–∏–π —Ç–µ—Å—Ç –ø–æ–¥–æ–±—Ä–∞—Ç—å —Ç—Ä—É–¥–Ω–æ. –í–æ—Å–ø–æ–ª—å–∑—É–µ–º—Å—è [–Ω–µ–ø–∞—Ä–∞–º–µ—Ç—Ä–∏—á–µ—Å–∫–∏–º —Ç–µ—Å—Ç–æ–º –£–∏–ª–∫–æ–∫—Å–æ–Ω–∞](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.wilcoxon.html) 

```python
wilcoxon_test_ultra_duration = st.wilcoxon(
    data_ultra.distance - 3130,
    alternative='greater'  
)

print(f'–¢–µ—Å—Ç –£–∏–ª–∫–æ–∫—Å–æ–Ω–∞:\
 p-value = {wilcoxon_test_ultra_duration.pvalue}')

alpha = 0.05

if wilcoxon_test_ultra_duration.pvalue < alpha:
    print('–ì–∏–ø–æ—Ç–µ–∑–∞ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è')
else:
    print('–ì–∏–ø–æ—Ç–µ–∑–∞ –Ω–µ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è')
```

–ü–æ–ø—Ä–æ–±—É–µ–º –∏ t-test

```python
ttest_ultra_duration = st.ttest_1samp(
    data_ultra.distance,
    3130,
    alternative='greater'
)

print(f'T-test:\
 p-value = {ttest_ultra_duration.pvalue}')

alpha = 0.05

if ttest_ultra_duration.pvalue < alpha:
    print('–ì–∏–ø–æ—Ç–µ–∑–∞ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è')
else:
    print('–ì–∏–ø–æ—Ç–µ–∑–∞ –Ω–µ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è')
```

```python
print(f'''–°—Ä–µ–¥–Ω—è—è –∏ –º–µ–¥–∏–∞–Ω–Ω–∞—è –¥–∞–ª—å–Ω–æ—Å—Ç—å –¥–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å –ø–æ–¥–ø–∏—Å–∫–æ–π:
    mean: {data_ultra.distance.mean()}
    median: {data_ultra.distance.median()}''')
```

### –ì–∏–ø–æ—Ç–µ–∑–∞ –æ –ø–æ–º–µ—Å—è—á–Ω–æ–π –≤—ã—Ä—É—á–∫–µ –¥–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å –ø–æ–¥–ø–∏—Å–∫–æ–π –∏ –±–µ–∑

–°—Ç–æ–∏—Ç –ø—Ä–æ–≤–µ—Ä–∏—Ç—å –≥–∏–ø–æ—Ç–µ–∑—É –æ —Ç–æ–º, –±—É–¥–µ—Ç –ª–∏ –ø–æ–º–µ—Å—è—á–Ω–∞—è –≤—ã—Ä—É—á–∫–∞ –æ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å –ø–æ–¥–ø–∏—Å–∫–æ–π –ø–æ –º–µ—Å—è—Ü–∞–º –≤—ã—à–µ, —á–µ–º –≤—ã—Ä—É—á–∫–∞ –æ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –±–µ–∑ –ø–æ–¥–ø–∏—Å–∫–∏. 

–°—Ñ–æ—Ä–º—É–ª–∏—Ä—É–µ–º –≥–∏–ø–æ—Ç–µ–∑—ã

 $H_0$: –°—Ä–µ–¥–Ω—è—è –ø–æ–º–µ—Å—è—á–Ω–∞—è –≤—ã—Ä—É—á–∫–∞ –æ—Ç –¥–≤—É—Ö –≥—Ä—É–ø–ø —Ä–∞–≤–Ω–∞

 $H_1$: –°—Ä–µ–¥–Ω—è—è –≤—ã—Ä—É—á–∫–∞ –æ—Ç –ø–æ–¥–ø–∏—Å—á–∏–∫–æ–≤ –≤—ã—à–µ

```python
data_revenue_ultra = (
    aggregated_data
    .loc[aggregated_data['subscription_type'] == 'ultra', 'revenue']
    )

data_revenue_free = (
    aggregated_data
    .loc[aggregated_data['subscription_type'] == 'free', 'revenue']
    )
```

–ü—Ä–æ–≤–µ—Ä–∏–º –∫–∞–∫ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –¥–∞–Ω–Ω—ã–µ

```python
shapiro_ultra_revenue = st.shapiro(data_revenue_ultra)
shapiro_free_revenue = st.shapiro(data_revenue_free)

print(f'''–¢–µ—Å—Ç –®–∞–ø–∏—Ä–æ-–£–∏–ª–∫–∞ –¥–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å –ø–æ–¥–ø–∏—Å–∫–æ–π:\
 p-value = {shapiro_ultra_revenue.pvalue}''')
print(f'–¢–µ—Å—Ç –®–∞–ø–∏—Ä–æ-–£–∏–ª–∫–∞ –¥–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –±–µ–∑ –ø–æ–¥–ø–∏—Å–∫–∏:\
 p-value = {shapiro_free_revenue.pvalue}');
```

–ü—Ä–æ–≤–µ—Ä–∏–º –∞—Å–∏–º–º–µ—Ç—Ä–∏—á–Ω–æ—Å—Ç—å –¥–∞–Ω–Ω—ã—Ö

```python
print(f'–ü–æ–∫–∞–∑–∞—Ç–µ–ª—å –ø–µ—Ä–µ–∫–æ—Å–∞ –≤—ã—Ä—É—á–∫–∏ –æ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å –ø–æ–¥–ø–∏—Å–∫–æ–π:\
 {data_revenue_ultra.skew()}')

print(f'–ü–æ–∫–∞–∑–∞—Ç–µ–ª—å –ø–µ—Ä–µ–∫–æ—Å–∞ –≤—ã—Ä—É—á–∫–∏ –æ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –±–µ–∑ –ø–æ–¥–ø–∏—Å–∫–∏:\
 {data_revenue_free.skew()}')
```

–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –¥–∞–Ω–Ω—ã–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –Ω–µ –Ω–æ—Ä–º–∞–ª—å–Ω–æ. –ü—Ä–∏ —ç—Ç–æ–º –µ—Å—Ç—å —Å–∏–ª—å–Ω—ã–π –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–π (–ø—Ä–∞–≤—ã–π) –ø–µ—Ä–µ–∫–æ—Å. –ë—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Ç–µ—Å—Ç –ú–∞–Ω–Ω–∞-–£–∏—Ç–Ω–∏

**–¢–µ—Å—Ç –ú–∞–Ω–Ω–∞-–£–∏—Ç–Ω–∏**

```python
u_test_revenue = st.mannwhitneyu(
    data_revenue_ultra,
    data_revenue_free,
    alternative='greater'
)

alpha = 0.05

print(f'U-test: p-value = {u_test_revenue.pvalue}')

if u_test_revenue.pvalue < alpha:
    print('–ì–∏–ø–æ—Ç–µ–∑–∞ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è')
else:
    print('–ì–∏–ø–æ—Ç–µ–∑–∞ –Ω–µ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è')
```

–¢–∞–∫–∂–µ –ø—Ä–æ–≤–µ–¥–µ–º –∏ t-test

```python
t_test_revenue = st.ttest_ind(
    data_revenue_ultra,
    data_revenue_free,
    alternative='greater'
    )

alpha = 0.05

print(f'–ù–µ–∑–∞–≤–∏—Å–∏–º—ã–π t-test: p-value = {t_test_revenue.pvalue}')

if t_test_revenue.pvalue < alpha:
    print('–ì–∏–ø–æ—Ç–µ–∑–∞ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è')
else:
    print('–ì–∏–ø–æ—Ç–µ–∑–∞ –Ω–µ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è')
```

```python
print(f'''–°—Ä–µ–¥–Ω—è—è –∏ –º–µ–¥–∏–∞–Ω–Ω–∞—è –≤—ã—Ä—É—á–∫–∞ –¥–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å –ø–æ–¥–ø–∏—Å–∫–æ–π:
    mean: {data_revenue_ultra.mean()}
    median: {data_revenue_ultra.median()}''')

print(f'''\n–°—Ä–µ–¥–Ω—è—è –∏ –º–µ–¥–∏–∞–Ω–Ω–∞—è –≤—ã—Ä—É—á–∫–∞ –¥–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –±–µ–∑ –ø–æ–¥–ø–∏—Å–∫–∏:
    mean: {data_revenue_free.mean()}
    median: {data_revenue_free.median()}''')
```

### –ó–∞–¥–∞—á–∞ 1

–û—Ç–¥–µ–ª—É –º–∞—Ä–∫–µ—Ç–∏–Ω–≥–∞ GoFast –ø–æ—Å—Ç–∞–≤–∏–ª–∏ –∑–∞–¥–∞—á—É: –Ω—É–∂–Ω–æ –ø—Ä–æ–≤–µ—Å—Ç–∏ –∞–∫—Ü–∏—é —Å —Ä–∞–∑–¥–∞—á–µ–π –ø—Ä–æ–º–æ–∫–æ–¥–æ–≤ –Ω–∞ –æ–¥–∏–Ω –±–µ—Å–ø–ª–∞—Ç–Ω—ã–π –º–µ—Å—è—Ü –ø–æ–¥–ø–∏—Å–∫–∏, –≤ —Ä–∞–º–∫–∞—Ö –∫–æ—Ç–æ—Ä–æ–π –∫–∞–∫ –º–∏–Ω–∏–º—É–º 
100 —Å—É—â–µ—Å—Ç–≤—É—é—â–∏—Ö –∫–ª–∏–µ–Ω—Ç–æ–≤ –¥–æ–ª–∂–Ω—ã –ø—Ä–æ–¥–ª–∏—Ç—å —ç—Ç—É –ø–æ–¥–ø–∏—Å–∫—É. –¢–æ –µ—Å—Ç—å –ø–æ –∑–∞–≤–µ—Ä—à–µ–Ω–∏–∏ –ø–µ—Ä–∏–æ–¥–∞ –¥–µ–π—Å—Ç–≤–∏—è –ø–æ–¥–ø–∏—Å–∫–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –º–æ–∂–µ—Ç –ª–∏–±–æ –æ—Ç–∫–∞–∑–∞—Ç—å—Å—è –æ—Ç –Ω–µ—ë, –ª–∏–±–æ –ø—Ä–æ–¥–ª–∏—Ç—å, —Å–æ–≤–µ—Ä—à–∏–≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–π –ø–ª–∞—Ç—ë–∂. 

–≠—Ç–∞ –∞–∫—Ü–∏—è —É–∂–µ –ø—Ä–æ–≤–æ–¥–∏–ª–∞—Å—å —Ä–∞–Ω–µ–µ –∏ –ø–æ –∏—Ç–æ–≥—É –≤—ã—è—Å–Ω–∏–ª–æ—Å—å, —á—Ç–æ –ø–æ—Å–ª–µ –±–µ—Å–ø–ª–∞—Ç–Ω–æ–≥–æ –ø—Ä–æ–±–Ω–æ–≥–æ –ø–µ—Ä–∏–æ–¥–∞ –ø–æ–¥–ø–∏—Å–∫—É –ø—Ä–æ–¥–ª–µ–≤–∞—é—Ç 10% –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π. –í—ã—è—Å–Ω–∏—Ç–µ, –∫–∞–∫–æ–µ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–º–æ–∫–æ–¥–æ–≤ –Ω—É–∂–Ω–æ —Ä–∞–∑–æ—Å–ª–∞—Ç—å, —á—Ç–æ–±—ã –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –Ω–µ –≤—ã–ø–æ–ª–Ω–∏—Ç—å –ø–ª–∞–Ω –±—ã–ª–∞ –ø—Ä–∏–º–µ—Ä–Ω–æ 5%. –ü–æ–¥–±–µ—Ä–∏—Ç–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è, –æ–ø–∏—Å—ã–≤–∞—é—â–µ–≥–æ —ç—Ç—É —Å–∏—Ç—É–∞—Ü–∏—é, –ø–æ—Å—Ç—Ä–æ–π—Ç–µ –≥—Ä–∞—Ñ–∏–∫ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∏ —Å—Ñ–æ—Ä–º—É–ª–∏—Ä—É–π—Ç–µ –æ—Ç–≤–µ—Ç –Ω–∞ –≤–æ–ø—Ä–æ—Å –æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ –ø—Ä–æ–º–æ–∫–æ–¥–æ–≤.

–î–ª—è —Ä–µ—à–µ–Ω–∏—è —ç—Ç–æ–π –∑–∞–¥–∞—á–∏ —Å—Ç–æ–∏—Ç –≤–æ—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å—Å—è –±–∏–Ω–æ–º–∏–∞–ª—å–Ω—ã–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ–º

–ò–º–µ–µ–º:

–¢—Ä–µ–±—É–µ–º–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É—Å–ø–µ—Ö–æ–≤: $k=100$

–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —É—Å–ø–µ—Ö–∞: $p=0.1$

–î–æ–ø—É—Å—Ç–∏–º–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –Ω–µ—É–¥–∞—á–∏ $P(k < 100) = 5\%$

–ë—É–¥–µ–º –ø–æ–¥–±–∏—Ä–∞—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ n, —á—Ç–æ–±—ã –≤ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–µ –≤—ã–ø–æ–ª–Ω—è–ª–æ—Å—å —É—Å–ª–æ–≤–∏–µ –Ω–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –Ω–µ—É–¥–∞—á–∏

```python
k = 100
p = 0.1

for n in range(100, 10000):
    
    distr = st.binom(n, p)

    if distr.cdf(k - 1) <= 0.05:
        print(f'–ü–æ–¥—Ö–æ–¥—è—â–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ n = {n}')
        break
    elif n == 10000:
        print('–ù–∞–¥–æ —É–≤–µ–ª–∏—á–∏—Ç—å –∑–æ–Ω—É –ø–æ–∏—Å–∫–∞')
```

–ú—ã –Ω–∞—à–ª–∏ –ø–æ–¥—Ö–æ–¥—è—â–∏–π –ø–∞—Ä–∞–º–µ—Ç—Ä. –¢–µ–ø–µ—Ä—å –º–æ–∂–Ω–æ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è.

```python
k_values = np.arange(distr.ppf(0.001), distr.ppf(0.999))

ax = sns.barplot(
    x=k_values,
    y=[distr.pmf(k) for k in k_values]
)

step = 10

ax.axvline(k / step, ls='--', label=f'k = {k}')
ax.xaxis.set_major_locator(MultipleLocator(step))

plt.title(f'–ë–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ (n={n}, p={p})')
plt.xlabel('–ß–∏—Å–ª–æ –ø—Ä–æ–¥–ª–µ–Ω–∏–π')
plt.ylabel('–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å')
plt.legend();
```

40 % –ø–æ–ª—É—á–∏–≤—à–∏—Ö –∫–ª–∏–µ–Ω—Ç–æ–≤. –û—Ç–¥–µ–ª –ø–ª–∞–Ω–∏—Ä—É–µ—Ç —Ä–∞–∑–æ—Å–ª–∞—Ç—å 1 –º–ª–Ω —É–≤–µ–¥–æ–º–ª–µ–Ω–∏–π. –° –ø–æ–º–æ—â—å—é –∞–ø–ø—Ä–æ–∫—Å–∏–º–∞—Ü–∏–∏ –ø–æ—Å—Ç—Ä–æ–π—Ç–µ –ø—Ä–∏–º–µ—Ä–Ω—ã–π –≥—Ä–∞—Ñ–∏–∫ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∏ –æ—Ü–µ–Ω–∏—Ç–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —É–≤–µ–¥–æ–º–ª–µ–Ω–∏–µ –æ—Ç–∫—Ä–æ—é—Ç –Ω–µ –±–æ–ª–µ–µ 399,5 —Ç—ã—Å. –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π.

–£ –Ω–∞—Å –µ—Å—Ç—å –±–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ:

$n = 10^6$

$p = 0.4$

–î–ª—è 1 –º–ª–Ω —É–≤–µ–¥–æ–º–ª–µ–Ω–∏–π –º—ã –º–æ–∂–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –Ω–æ—Ä–º–∞–ª—å–Ω—É—é –∞–ø–ø—Ä–æ–∫—Å–∏–º–∞—Ü–∏—é —Å–æ —Å–ª–µ–¥—É—é—â–∏–º–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏:

$\mu = n \cdot p = 10^6 \cdot 0.4 = 400000$

$\sigma = \sqrt{n \cdot p \cdot (1-p)} = \sqrt{10^6 \cdot 0.4 \cdot 0.6} = \sqrt{240000}$

–ù–∞–π–¥–µ–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —É–≤–µ–¥–æ–º–ª–µ–Ω–∏–µ –æ—Ç–∫—Ä–æ—é—Ç –Ω–µ –±–æ–ª–µ–µ 399.5 —Ç—ã—Å. –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π

```python
n = 1e6
p = 0.4

mu = n * p
sigma = np.sqrt(n * p * (1 - p))

distr = st.norm(mu, sigma)

print(f'–ò—Å–∫–æ–º–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å P = {distr.cdf(399500):.2%}')
```

–ü–æ—Å—Ç—Ä–æ–∏–º –ø—Ä–∏–±–ª–∏–∑–∏—Ç–µ–ª—å–Ω—ã–π –≥—Ä–∞—Ñ–∏–∫ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è

```python
x = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 1000)

y = distr.pdf(x)

sns.lineplot(
    x=x, 
    y=y
)

plt.axvline(399500, ls='--', label=f'k = 399500')

plt.title(f'–ù–æ—Ä–º–∞–ª—å–Ω–∞—è –∞–ø–ø—Ä–æ–∫—Å–∏–º–∞—Ü–∏—è –±–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è \
(n={int(n)}, p={p})')

plt.xlabel('–ß–∏—Å–ª–æ –æ—Ç–∫—Ä—ã—Ç–∏–π')
plt.ylabel('–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å')
plt.legend();
```
